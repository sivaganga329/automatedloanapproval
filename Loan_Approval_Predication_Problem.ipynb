```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("Dataset.csv")  # Ensure the correct path

# Convert categorical columns to numerical using encoding (if needed)
df_numeric = df.select_dtypes(include=['number'])  # Select only numeric columns

# Compute correlation
corr = df_numeric.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='BuPu')
plt.title("Correlation Matrix")
plt.show()
```

```python
## Total Applicant Income

df['Total_Income'] = df['ApplicantIncome'] + df['CoapplicantIncome']
df.head()
```

```python
## Apply Log Transformation

df['ApplicantIncomelog'] = np.log(df['ApplicantIncome'] + 1)
sns.distplot(df['ApplicantIncomelog'])
```

```python
df.head()
```

```python

df['LoanAmountlog'] = np.log(df['LoanAmount'] + 1)
sns.distplot(df['LoanAmountlog'])
```

```python

df['Loan_Amount_Term_log'] = np.log(df['Loan_Amount_Term'] + 1)
sns.distplot(df['Loan_Amount_Term_log'])
```

```python

df['Total_Income_log'] = np.log(df['Total_Income'] + 1)
sns.distplot(df['Total_Income_log'])
```

```python
df.head()
```

```python
## drop unnecessary columns
cols = ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Total_Income','Loan_ID']
df = df.drop(columns = cols, axis = 1)
df.head()
```

```python
## Encoding Technique : Label Encoding, One Hot Encoding

from sklearn.preprocessing import LabelEncoder
cols = ['Gender','Married','Education','Dependents','Self_Employed','Property_Area','Loan_Status']
le =  LabelEncoder()
for col in cols:
  df[col] =  le.fit_transform(df[col])
```

```python
df.head()
```

```python
df.dtypes
```

```python
## Split Independent and dependent features

X = df.drop(columns = ['Loan_Status'],axis = 1)
y = df['Loan_Status']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25,random_state = 42)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Handle missing values using mean (for numerical) or most frequent (for categorical)
imputer = SimpleImputer(strategy="mean")  # You can change strategy to "median" or "most_frequent" if needed
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Train Logistic Regression model
model1 = LogisticRegression()
model1.fit(X_train_imputed, y_train)

# Predict and evaluate
y_pred_model1 = model1.predict(X_test_imputed)
accuracy = accuracy_score(y_test, y_pred_model1)

print("Accuracy score of Logistic Regression:", accuracy * 100)
```
